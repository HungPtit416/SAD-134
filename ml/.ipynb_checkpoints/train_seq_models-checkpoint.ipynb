## Train RNN / LSTM / BiLSTM (Next-action classification)

Notebook này dùng `data/data_user500.csv` để:
- Tạo tập dữ liệu dự đoán **hành vi kế tiếp** (8 lớp)
- Train 3 mô hình: **RNN**, **LSTM**, **BiLSTM**
- Đánh giá (Accuracy, Macro-F1), vẽ **learning curves** + **confusion matrix**
- Chọn mô hình tốt nhất `model_best` theo **Macro-F1**
- Lưu artifacts vào `ml/artifacts/`

> Bạn chạy theo thứ tự từ trên xuống (hoặc Run All).

```python
# If you don't have the ML environment yet, run:
# !pip install -r ml/requirements.txt

import os
from pathlib import Path

import numpy as np
import pandas as pd

# Reproducibility
SEED = 42
np.random.seed(SEED)

DATA_CSV = Path('data') / 'data_user500.csv'
OUT_DIR = Path('ml') / 'artifacts'
PLOTS_DIR = OUT_DIR / 'plots'
OUT_DIR.mkdir(parents=True, exist_ok=True)
PLOTS_DIR.mkdir(parents=True, exist_ok=True)

print('CSV exists:', DATA_CSV.exists(), str(DATA_CSV))
print('Artifacts dir:', OUT_DIR.resolve())
```

```python
BEHAVIORS_8 = [
    'view',
    'click',
    'add_to_cart',
    'purchase',
    'search',
    'browse_products',
    'browse_recommended',
    'checkout',
]

label_to_id = {b: i for i, b in enumerate(BEHAVIORS_8)}
id_to_label = {i: b for b, i in label_to_id.items()}

df = pd.read_csv(DATA_CSV)
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce', utc=True)
df = df.dropna(subset=['user_id', 'action', 'timestamp']).copy()
df = df[df['action'].isin(label_to_id)].copy()
df['user_id'] = df['user_id'].astype(str)
df['label_id'] = df['action'].map(label_to_id).astype(int)
df = df.sort_values(['user_id', 'timestamp'], ascending=True)

print('rows:', len(df))
print('users:', df['user_id'].nunique())
print('actions:', df['action'].value_counts().to_dict())
```

```python
from sklearn.model_selection import train_test_split

SEQ_LEN = 6

# Split by user to avoid leakage
users = df['user_id'].drop_duplicates().tolist()
u_train, u_tmp = train_test_split(users, test_size=0.30, random_state=SEED, shuffle=True)
u_val, u_test = train_test_split(u_tmp, test_size=0.50, random_state=SEED, shuffle=True)


def to_xy(uids):
    x_list = []
    y_list = []
    for uid, g in df[df['user_id'].isin(uids)].groupby('user_id', sort=False):
        seq = g['label_id'].tolist()
        if len(seq) <= SEQ_LEN:
            continue
        for i in range(SEQ_LEN, len(seq)):
            x_list.append(seq[i-SEQ_LEN:i])
            y_list.append(seq[i])
    return np.asarray(x_list, dtype=np.int32), np.asarray(y_list, dtype=np.int32)

x_train, y_train = to_xy(set(u_train))
x_val, y_val = to_xy(set(u_val))
x_test, y_test = to_xy(set(u_test))

print('x_train:', x_train.shape, 'x_val:', x_val.shape, 'x_test:', x_test.shape)
assert x_train.shape[0] > 1000, 'Not enough training samples. Increase events per user or reduce SEQ_LEN.'
```

```python
import tensorflow as tf

tf.random.set_seed(SEED)


def build_model(kind: str, *, vocab_size: int, seq_len: int, embed_dim: int, rnn_units: int, num_classes: int):
    inputs = tf.keras.Input(shape=(seq_len,), dtype='int32')
    x = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embed_dim, name='action_embedding')(inputs)

    if kind == 'rnn':
        x = tf.keras.layers.SimpleRNN(rnn_units, name='rnn')(x)
    elif kind == 'lstm':
        x = tf.keras.layers.LSTM(rnn_units, name='lstm')(x)
    elif kind == 'bilstm':
        x = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(rnn_units), name='bilstm')(x)
    else:
        raise ValueError(f'Unsupported model kind: {kind}')

    x = tf.keras.layers.Dense(64, activation='relu')(x)
    x = tf.keras.layers.Dropout(0.2)(x)
    outputs = tf.keras.layers.Dense(num_classes, activation='softmax')(x)

    model = tf.keras.Model(inputs=inputs, outputs=outputs, name=f'{kind}_next_action')
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy'],
    )
    return model


EMBED_DIM = 16
RNN_UNITS = 64
EPOCHS = 10
BATCH_SIZE = 256

print('TensorFlow:', tf.__version__)
```

```python
import json
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score, f1_score


def plot_curves(history, name: str):
    plt.figure(figsize=(9, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history.get('loss', []), label='train')
    plt.plot(history.history.get('val_loss', []), label='val')
    plt.title(f'{name} loss')
    plt.xlabel('epoch')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history.get('accuracy', []), label='train')
    plt.plot(history.history.get('val_accuracy', []), label='val')
    plt.title(f'{name} accuracy')
    plt.xlabel('epoch')
    plt.legend()

    plt.tight_layout()
    path = PLOTS_DIR / f'{name}_curves.png'
    plt.savefig(path, dpi=150)
    plt.show()
    plt.close()
    return path


def plot_confusion(y_true, y_pred, name: str):
    fig, ax = plt.subplots(figsize=(7, 6))
    ConfusionMatrixDisplay.from_predictions(
        y_true,
        y_pred,
        display_labels=[id_to_label[i] for i in range(len(BEHAVIORS_8))],
        xticks_rotation=45,
        colorbar=False,
        normalize='true',
        ax=ax,
    )
    ax.set_title(f'{name} confusion matrix (normalized)')
    fig.tight_layout()
    path = PLOTS_DIR / f'{name}_confusion.png'
    fig.savefig(path, dpi=150)
    plt.show()
    plt.close(fig)
    return path


results = {}
best_name = None
best_macro_f1 = -1.0
best_model = None

for kind in ['rnn', 'lstm', 'bilstm']:
    model = build_model(
        kind,
        vocab_size=len(BEHAVIORS_8),
        seq_len=SEQ_LEN,
        embed_dim=EMBED_DIM,
        rnn_units=RNN_UNITS,
        num_classes=len(BEHAVIORS_8),
    )

    cb = [tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)]
    history = model.fit(
        x_train,
        y_train,
        validation_data=(x_val, y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        verbose=2,
        callbacks=cb,
    )

    probs = model.predict(x_test, batch_size=512, verbose=0)
    y_pred = probs.argmax(axis=1).astype(np.int32)

    acc = float(accuracy_score(y_test, y_pred))
    macro_f1 = float(f1_score(y_test, y_pred, average='macro'))
    results[kind] = {'accuracy': acc, 'macro_f1': macro_f1}

    print('\n===', kind, '===')
    print('accuracy:', acc)
    print('macro_f1:', macro_f1)

    plot_curves(history, kind)
    plot_confusion(y_test, y_pred, kind)

    if macro_f1 > best_macro_f1:
        best_macro_f1 = macro_f1
        best_name = kind
        best_model = model

# Save best model + results
meta = {
    'task': 'next_action_prediction',
    'label_space': BEHAVIORS_8,
    'seq_len': int(SEQ_LEN),
    'embed_dim': int(EMBED_DIM),
    'rnn_units': int(RNN_UNITS),
    'metrics': results,
    'model_best': best_name,
    'selection_criterion': 'macro_f1',
}

(OUT_DIR / 'results.json').write_text(json.dumps(meta, ensure_ascii=False, indent=2), encoding='utf-8')
(OUT_DIR / 'label_map.json').write_text(json.dumps(label_to_id, ensure_ascii=False, indent=2), encoding='utf-8')

best_path = OUT_DIR / 'model_best.keras'
best_model.save(best_path)

print('\nSaved:', best_path)
print(json.dumps(meta, ensure_ascii=False, indent=2))
```
